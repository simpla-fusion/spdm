```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.constants
from spdm.core.Expression import Variable
from spdm.core.Function import Function
```

```python
x = np.linspace(0, scipy.constants.pi * 2.0, 16)
f = Function(x, np.sin(x), name="f")
g = Function(x, np.cos(x), name="g")
X = Variable(0, "X")
```

```python
g(X)
```

```python
plt.plot(x,f(x),label="f")
plt.plot(x,f.I(x),label="I")
plt.plot(x,f.d(x),label="d")
plt.legend()
```

```python
F = f(g)
```

```python
F(x)
```

```python
dF0 = f.d(g) + f * g.d
dF1 = F.d
dF1(x) - dF0(x)
```

```python
plt.plot(x,F(x) ,label="F")
plt.plot(x,dF1(x),label="dF")
plt.plot(x,np.sin(np.cos(x)) ,label="sin(cos(x))")
plt.plot(x,-np.cos(np.cos(x))*np.sin(x) ,label="d[sin(cos(x))]")
plt.plot(x,f.d(g(x))*g.d(x),label="dF0")

plt.legend()
```

```python
plt.plot(x,f.d(g(x))+f(g(x))*g.d(x),label="dF0")
plt.plot(x,f(g(x)) ,label="f(g)")
```
